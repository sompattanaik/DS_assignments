# Q-1

```python
import pandas as pd
import numpy as np
import scipy
from scipy import stats
```

```python
cutlet = pd.read_csv('Cutlets.csv')
cutlet
```

```python
cutlet.describe()
```

```python
unit_a = pd.Series(cutlet.iloc[:,0])
unit_a
```

```python
unit_b = pd.Series(cutlet.iloc[:,1])
unit_b
```

# As the manager wants to compare between 2 samples, it is a 2-sample t test.

```python
t_stats,p_value = stats.ttest_ind(unit_a,unit_b)
t_stats,p_value
```

```python
print('p_value for the 2 sample test is : ', p_value,)
```

#The p-value < 0.05 so we can accept the Altenate Hypothesis
